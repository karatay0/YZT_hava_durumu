```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

file_path = 'munich.csv' 
df = pd.read_csv(file_path, sep=';')

df['time'] = pd.to_datetime(df['time'])

df.set_index('time', inplace=True)

df.fillna(0, inplace=True)

print("--- Veri Hazır (İlk 5 Satır) ---")
df.head()
```

```python
print("--- Temel İstatistikler ---")
df.describe()
```

```python
# --- Genel Toplamlar ---
# Gözlem süresi boyunca toplam yağış ve kar miktarı
toplam_yagis = df['precipitation_sum (mm)'].sum()
toplam_kar = df['snowfall_sum (cm)'].sum()

print(f"Gözlem Süresi Boyunca Toplam Yağış: {toplam_yagis:.2f} mm")
print(f"Gözlem Süresi Boyunca Toplam Kar Yağışı: {toplam_kar:.2f} cm")

# --- Gün Sayıları ---
yagisli_gun_sayisi = (df['precipitation_sum (mm)'] > 0).sum()
karli_gun_sayisi = (df['snowfall_sum (cm)'] > 0).sum()

print(f"\nYağış Kaydedilen Gün Sayısı: {yagisli_gun_sayisi}")
print(f"Kar Kaydedilen Gün Sayısı: {karli_gun_sayisi}")

en_yagisli_gun_tarihi = df['precipitation_sum (mm)'].idxmax()
en_yüksek_yagis_miktari = df['precipitation_sum (mm)'].max()

print(f"\nEn Yüksek Yağış Olan Gün: {en_yagisli_gun_tarihi.strftime('%Y-%m-%d')} (Miktar: {en_yüksek_yagis_miktari} mm)")

# En yüksek kar yağışının olduğu gün (eğer kar yağdıysa)
if karli_gun_sayisi > 0:
    en_karli_gun_tarihi = df['snowfall_sum (cm)'].idxmax()
    en_yüksek_kar_miktari = df['snowfall_sum (cm)'].max()
    print(f"En Yüksek Kar Olan Gün: {en_karli_gun_tarihi.strftime('%Y-%m-%d')} (Miktar: {en_yüksek_kar_miktari} cm)")
else:
    print("Veri setinde kar yağışı kaydedilmemiş.")
```

```python

aylik_toplamlar = df.resample('ME').sum()

print("--- Aylık Toplam Yağış ve Kar (mm/cm) ---")
print(aylik_toplamlar)

aylik_toplam_yagis = aylik_toplamlar['precipitation_sum (mm)']

en_yagisli_ay = aylik_toplam_yagis.idxmax()
en_yüksek_aylik_yagis = aylik_toplam_yagis.max()

print(f"\nEn Yağışlı Ay (Toplam): {en_yagisli_ay.strftime('%Y-%m')} (Toplam Miktar: {en_yüksek_aylik_yagis:.2f} mm)")

en_kurak_ay = aylik_toplam_yagis.idxmin()
en_düsük_aylik_yagis = aylik_toplam_yagis.min()

print(f"En Kurak Ay (Toplam): {en_kurak_ay.strftime('%Y-%m')} (Toplam Miktar: {en_düsük_aylik_yagis:.2f} mm)")
```

```python
# --- Grafik 1: Günlük Yağış Miktarı Zaman Serisi ---
plt.figure(figsize=(15, 6))

plt.plot(df.index, df['precipitation_sum (mm)'], label='Günlük Yağış', color='blue', alpha=0.7)

plt.title('Münih Günlük Yağış Miktarı (Mart - ... 2024)')
plt.xlabel('Tarih')
plt.ylabel('Yağış Miktarı (mm)')

plt.legend()

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(12, 6))

aylik_indeks_str = aylik_toplamlar.index.strftime('%Y-%m')
plt.bar(aylik_indeks_str, aylik_toplamlar['precipitation_sum (mm)'], color='teal', width=0.6)

plt.title('Aylık Toplam Yağış Miktarı - Münih 2024')
plt.xlabel('Ay')
plt.ylabel('Toplam Yağış (mm)')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
yagisli_gunler = df[df['precipitation_sum (mm)'] > 0]['precipitation_sum (mm)']

plt.figure(figsize=(10, 6))

plt.hist(yagisli_gunler, bins=20, color='darkred', edgecolor='black')

plt.title('Yağışlı Günlerdeki Yağış Miktarı Dağılımı (Histogram)')
plt.xlabel('Yağış Miktarı (mm)')
plt.ylabel('Gün Sayısı (Frekans)')
plt.yscale('log')
                 
plt.show()
```
